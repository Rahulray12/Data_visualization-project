```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_url = "/kaggle/input/indian-liver-patient-records/indian_liver_patient.csv"
df = pd.read_csv(data_url)
print(df.head())
print(df.info())
print(df.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Dataset', data=df)
plt.title('Distribution of Liver Disease')
plt.xlabel('Liver Disease')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Dataset', y='Age', data=df)
plt.title('Age vs Liver Disease')
plt.xlabel('Liver Disease')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Dataset', y='Total_Bilirubin', data=df)
plt.title('Total Bilirubin vs Liver Disease')
plt.xlabel('Liver Disease')
plt.ylabel('Total Bilirubin')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Dataset', y='Alkaline_Phosphotase', data=df)
plt.title('Alkaline Phosphotase vs Liver Disease')
plt.xlabel('Liver Disease')
plt.ylabel('Alkaline Phosphotase')
plt.show()
```
