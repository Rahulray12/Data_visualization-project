```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("/kaggle/input/students-data-analysis/Students data.csv")

# Check the column names
print(data.columns)

# Based on the column names, replace "math score" with the correct column name
# For example, if the column name is "Math_Score", use:
# sns.histplot(data=data, x="Math_Score", bins=30, kde=True)

# Now, let's start with some basic visualizations

# 1. Countplot of gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="gender", data=data)
plt.title("Gender Distribution")
plt.show()







```

```python
# 2. Histogram of math scores (update the column name)
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x="Functional_analysis", bins=30, kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
# 3. Boxplot of math scores by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x="gender", y="Functional_analysis", data=data)
plt.title("Math Score by Gender")
plt.show()
```
